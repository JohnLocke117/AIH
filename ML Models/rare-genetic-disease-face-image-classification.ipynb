```python
import os
import pandas as pd

# Path to the folder containing images
image_folder = "/kaggle/input/image-dataset/clf_model_images"  # Update with your folder path

# Lists of image numbers for each class
class_lists = extracted_image_ids

# Create a dictionary for quick lookup of class by image number
image_to_class = {}
for class_id, image_list in enumerate(class_lists):
    for image_number in image_list:
        image_to_class[image_number] = class_id

# Extract image paths and their class IDs
data = []
for filename in os.listdir(image_folder):
    if filename.endswith("_aligned.jpg"):  # Filter for relevant images
        # Extract the image number from the filename
        image_number = int(filename.split("_")[0])
        # Determine the class ID
        class_id = image_to_class.get(image_number, None)  # None if not found
        if class_id is not None:
            # Full path to the image
            image_path = os.path.join(image_folder, filename)
            # Append to the dataset
            data.append({"image": image_path, "class_id": class_id})

# Create a DataFrame

dataset = pd.DataFrame(data)

# Display the dataset
print(dataset.head())
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
from sklearn.model_selection import train_test_split
import torch
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, models
from torch import nn, optim
```

```python
df = dataset

# Display basic info
print(df.info())
print(df.head())

# Class distribution
class_counts = df['class_id'].value_counts()
print("Class Distribution:\n", class_counts)

# Plot class distribution
class_counts.plot(kind='bar', title='Class Distribution')
plt.show()
```

```python
train_df, val_df = train_test_split(df, test_size=0.2, stratify=df['class_id'], random_state=42)
print(f"Train size: {len(train_df)}, Validation size: {len(val_df)}")
```

```python
class ImageDataset(Dataset):
    def __init__(self, dataframe, transform=None):
        self.dataframe = dataframe
        self.transform = transform

    def __len__(self):
        return len(self.dataframe)

    def __getitem__(self, idx):
        row = self.dataframe.iloc[idx]
        image = Image.open(row['image']).convert('RGB')
        label = row['class_id']
        if self.transform:
            image = self.transform(image)
        return image, label
```

```python
# Transforms for data augmentation and normalization
train_transforms = transforms.Compose([
    transforms.Resize((128, 128)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(15),
    transforms.ToTensor(),
    transforms.Normalize([0.5, 0.5, 0.5], [0.5, 0.5, 0.5])
])

val_transforms = transforms.Compose([
    transforms.Resize((128, 128)),
    transforms.ToTensor(),
    transforms.Normalize([0.5, 0.5, 0.5], [0.5, 0.5, 0.5])
])

# Create datasets
train_dataset = ImageDataset(train_df, transform=train_transforms)
val_dataset = ImageDataset(val_df, transform=val_transforms)

# Create dataloaders
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)
```

```python
# Load pre-trained ResNet18
model = models.resnet18(pretrained=True)

# Modify the final layer for the number of classes
num_classes = df['class_id'].nunique()
model.fc = nn.Linear(model.fc.in_features, num_classes)

# Use GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

# 10 epoch Results

```python
def train_model(model, train_loader, val_loader, criterion, optimizer, num_epochs=10):
    best_accuracy = 0.0

    for epoch in range(num_epochs):
        # Training phase
        model.train()
        train_loss = 0.0
        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)
            
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            train_loss += loss.item()

        # Validation phase
        model.eval()
        val_loss = 0.0
        correct = 0
        total = 0
        with torch.no_grad():
            for images, labels in val_loader:
                images, labels = images.to(device), labels.to(device)
                outputs = model(images)
                loss = criterion(outputs, labels)
                val_loss += loss.item()
                
                _, predicted = torch.max(outputs, 1)
                total += labels.size(0)
                correct += (predicted == labels).sum().item()

        accuracy = correct / total
        print(f"Epoch {epoch+1}/{num_epochs}, Train Loss: {train_loss/len(train_loader):.4f}, "
              f"Val Loss: {val_loss/len(val_loader):.4f}, Val Accuracy: {accuracy:.4f}")

        # Save the best model
        if accuracy > best_accuracy:
            best_accuracy = accuracy
            torch.save(model.state_dict(), 'best_model.pth')
            print("Model saved!")
```

```python
train_model(model, train_loader, val_loader, criterion, optimizer, num_epochs=10)
```

```python
def visualize_feature_maps(model, image_path):
    model.eval()
    image = Image.open(image_path).convert('RGB')
    transform = transforms.Compose([
        transforms.Resize((128, 128)),
        transforms.ToTensor(),
    ])
    image_tensor = transform(image).unsqueeze(0).to(device)

    # Hook for intermediate layer
    features = []

    def hook_fn(module, input, output):
        features.append(output)

    layer = model.layer1  # Example: First layer
    hook = layer.register_forward_hook(hook_fn)

    # Forward pass
    model(image_tensor)
    hook.remove()

    # Plot feature maps
    feature_map = features[0].squeeze(0).detach().cpu().numpy()
    plt.figure(figsize=(12, 12))
    for i in range(min(16, feature_map.shape[0])):  # Plot first 16 channels
        plt.subplot(4, 4, i + 1)
        plt.imshow(feature_map[i], cmap='viridis')
        plt.axis('off')
    plt.show()

# Example usage
visualize_feature_maps(model, "/kaggle/input/image-dataset/clf_model_images/10001_aligned.jpg")
```

# 20 epoch Results

```python
def train_model(model, train_loader, val_loader, criterion, optimizer, num_epochs=10):
    best_accuracy = 0.0
    train_accuracies = []
    val_accuracies = []
    train_losses = []
    val_losses = []

    for epoch in range(num_epochs):
        # Training phase
        model.train()
        train_loss = 0.0
        correct_train = 0
        total_train = 0

        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)
            
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            train_loss += loss.item()

            # Track training accuracy
            _, predicted = torch.max(outputs, 1)
            total_train += labels.size(0)
            correct_train += (predicted == labels).sum().item()

        train_accuracy = correct_train / total_train
        train_losses.append(train_loss / len(train_loader))
        train_accuracies.append(train_accuracy)

        # Validation phase
        model.eval()
        val_loss = 0.0
        correct_val = 0
        total_val = 0

        with torch.no_grad():
            for images, labels in val_loader:
                images, labels = images.to(device), labels.to(device)
                outputs = model(images)
                loss = criterion(outputs, labels)
                val_loss += loss.item()

                # Track validation accuracy
                _, predicted = torch.max(outputs, 1)
                total_val += labels.size(0)
                correct_val += (predicted == labels).sum().item()

        val_accuracy = correct_val / total_val
        val_losses.append(val_loss / len(val_loader))
        val_accuracies.append(val_accuracy)

        print(f"Epoch {epoch+1}/{num_epochs}, Train Loss: {train_losses[-1]:.4f}, "
              f"Val Loss: {val_losses[-1]:.4f}, Train Accuracy: {train_accuracies[-1]:.4f}, "
              f"Val Accuracy: {val_accuracies[-1]:.4f}")

        # Save the best model
        if val_accuracy > best_accuracy:
            best_accuracy = val_accuracy
            torch.save(model.state_dict(), 'best_model2.pth')
            print("Model saved!")

    # Plot Accuracy Curves
    epochs = range(1, num_epochs + 1)
    plt.figure(figsize=(10, 6))
    plt.plot(epochs, train_accuracies, label='Training Accuracy', color='blue', marker='o')
    plt.plot(epochs, val_accuracies, label='Validation Accuracy', color='green', marker='o')
    plt.title('Accuracy Curve')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid()
    plt.show()

    # Plot Loss Curves
    plt.figure(figsize=(10, 6))
    plt.plot(epochs, train_losses, label='Training Loss', color='red', marker='o')
    plt.plot(epochs, val_losses, label='Validation Loss', color='orange', marker='o')
    plt.title('Loss Curve')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid()
    plt.show()
```
